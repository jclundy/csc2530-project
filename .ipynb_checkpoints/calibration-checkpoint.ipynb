```python
import numpy as np
import argparse
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image

import os
```

```python
image_size = (1920, 1080) # width, height
calibration_file = "calibration.npz"
test_image_file = ""


calibrationParams = np.load(calibration_file)
mtx = calibrationParams['mtx']
dist = calibrationParams['dist']
```

```python
w = image_size[0]
h = image_size[1]

# beta = 
newcameramtx, roi = cv.getOptimalNewCameraMatrix(mtx, dist, (w,h), 1, (w,h))
```

```python
base_folder = "C:/Users/littl/OneDrive - University of Toronto/Winter 2025/CSC2530/Project/datasets/distance_tests/distance_calibration"

numCaptures = 7
laserLocationPixelsRectified = np.zeros((numCaptures, 2))

image_type = "high" # high exposure
for i in range(1,numCaptures+1):
    print(i)
    imageName = image_type +".png"
    imageFile = os.path.join(base_folder, str(i), imageName)
    print(imageFile)
    img = cv.imread(imageFile)
    img = cv.cvtColor(img, cv.COLOR_RGB2BGR )
    plt.figure(figsize=(9, 6))
    # title = "image" +str(i)
    # plt.title(title)
    # plt.imshow(img)

    img_rectified = cv.undistort(img, mtx, dist, None, newcameramtx)
    plt.figure(figsize=(9, 6))
    # title = "image" +str(i) + "- rectified"
    # plt.title(title)
    # plt.imshow(img_rectified)

    gray_rectified = cv.cvtColor(img_rectified, cv.COLOR_BGR2GRAY)
    gray_rectified.shape
    # plt.figure(figsize=(9, 6))
    # plt.title(f"frame 0 - gray")
    # plt.imshow(gray_rectified, cmap='gray')

    radius = 7
    grayBlurred = cv.GaussianBlur(gray_rectified, (radius, radius), 0)
    (minVal, maxVal, minLoc, maxLoc) = cv.minMaxLoc(grayBlurred)

    laserLocationPixelsRectified[i-1][0] = maxLoc[0]
    laserLocationPixelsRectified[i-1][1] = maxLoc[1]    

    annotated = img_rectified.copy()
    cv.circle(annotated, maxLoc, radius, (0, 0, 255), 2)
    plt.figure(figsize=(9, 6))
    title = "image" +str(i) + "laser dot region"
    plt.title(title)
    plt.imshow(annotated)
```
